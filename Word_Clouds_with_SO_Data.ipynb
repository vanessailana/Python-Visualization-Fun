```python
import pandas as pd
import numpy as np

valid=pd.read_csv('/valid 2.csv')
train=pd.read_csv('/train.csv')
```

```python
high = train[train['Y']==0]
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['encoded']=le.fit_transform(train.Y)

valid['encoded']=le.fit_transform(valid.Y)

print(train)
```

```python
high= train[train['encoded']==0]
low_closed= train[train['encoded']==1]
low_opened= train[train['encoded']==2]
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib as plt
import matplotlib.pyplot as plt
plt.title("Trainset-High Quality Questions")
wc = WordCloud(max_words = 1000 , width = 1600 , height = 800 , stopwords = STOPWORDS).generate(" ".join(high['Body']))
plt.imshow(wc , interpolation = 'bilinear')
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib as plt
import matplotlib.pyplot as plt
plt.title("Trainset-Low Quality Closed Questions")
wc = WordCloud(max_words = 1000 , width = 1600 , height = 800 , stopwords = STOPWORDS).generate(" ".join(low_closed['Body']))
plt.imshow(wc , interpolation = 'bilinear')
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib as plt
import matplotlib.pyplot as plt
plt.title("Trainset-Low Quality Edit Questions")
wc = WordCloud(max_words = 1000 , width = 1600 , height = 800 , stopwords = STOPWORDS).generate(" ".join(low_opened['Body']))
plt.imshow(wc , interpolation = 'bilinear')
```
